```python
mnogochlen_2 = [[1,1,1],[1,8,8**2],[1,15,15**2]]
matrix_A2 = np.array(mnogochlen_2)
f_1 = math.sin(1 / 5) * math.exp(1 / 10) + 5 * math.exp(-1 / 2)
f_2 = math.sin(8 / 5) * math.exp(8 / 10) + 5 * math.exp(-8 / 2)
f_3 = math.sin(15 / 5) * math.exp(15 / 10) + 5 * math.exp(-15 / 2)
matrix_B2 = [f_1,f_2,f_3]
matrix_B2 = np.array(matrix_B2)
```
